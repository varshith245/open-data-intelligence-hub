# Data Cleaning Summary

The dataset initially contained no missing values in most columns. However, the TotalCharges column was stored as an object datatype due to 22 blank entries. These blank values were converted to NaN and replaced using the median value of the column. The customerID column was removed because it serves only as an identifier and does not contribute to churn analysis. Duplicate records were checked and handled appropriately.


# EDA

1.Churn Count Plot

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Churn', data=df)

plt.title('Customer Churn Distribution')
plt.show()
```

The number of customers who did not churn is significantly higher than those who churned. This indicates that most customers remained with the company, while a smaller proportion discontinued the service.


2.Churn Percentage Pie Chart

```python
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title('Customer Churn Percentage')
plt.ylabel('')
plt.show()
```

Approximately one-fourth of customers churned, while the majority continued using the company's services.

3.Tenure Histogram

```python
plt.figure(figsize=(8,5))

sns.histplot(df['tenure'], bins=20, kde=True)

plt.title('Distribution of Customer Tenure')
plt.xlabel('Tenure (Months)')
plt.show()
```

The tenure distribution shows that many customers are either very new or have stayed with the company for a long period. Long-tenure customers generally indicate better retention.

4. Boxplot

```python
plt.figure(figsize=(8,5))

sns.boxplot(y=df['MonthlyCharges'])

plt.title('Monthly Charges Boxplot')
plt.show()
```

The boxplot helps identify extreme values in MonthlyCharges. A few customers have unusually high monthly charges, indicating potential outliers.

5. Contract Type vs Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Contract',
    hue='Churn',
    data=df
)

plt.title('Contract Type vs Churn')
plt.xticks(rotation=15)
plt.show()
```

Customers with month-to-month contracts exhibit the highest churn rate. Customers with one-year and two-year contracts are more likely to remain with the company.

6.Internet Service vs Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='InternetService',
    hue='Churn',
    data=df
)

plt.title('Internet Service vs Churn')
plt.show()
```

Customers using fiber optic internet services show a higher churn rate compared to DSL users and customers without internet service.

7. Tenure vs Churn

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Churn',
    y='tenure',
    data=df
)

plt.title('Tenure vs Churn')
plt.show()
```

Churned customers generally have lower tenure compared to retained customers. This suggests that newer customers are more likely to leave.

8.Correlation Heatmap

```python
df['Churn_numeric'] = df['Churn'].map({
    'Yes':1,
    'No':0
})
```

```python
plt.figure(figsize=(8,6))

corr = df[['SeniorCitizen',
           'tenure',
           'MonthlyCharges',
           'TotalCharges',
           'Churn_numeric']].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```

Tenure and TotalCharges show strong positive correlation. Churn has a negative relationship with tenure, indicating that long-term customers are less likely to churn.

9. Scatter Plot

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='MonthlyCharges',
    y='TotalCharges',
    hue='Churn',
    data=df
)

plt.title('Monthly Charges vs Total Charges')
plt.show()
```

Customers with high total charges generally have longer tenure. Churned customers are concentrated among customers with lower total charges and shorter service duration.

```python
df[['tenure','MonthlyCharges','TotalCharges']].describe()
```

The average customer tenure is around 32 months. Monthly charges vary widely among customers, indicating different subscription plans. Total charges exhibit a large range due to differences in tenure and service usage.
