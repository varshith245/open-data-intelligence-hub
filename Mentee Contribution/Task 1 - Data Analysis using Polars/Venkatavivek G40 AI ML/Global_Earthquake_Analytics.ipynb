```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python

df=pd.read_csv("earthquake_data.csv")
df
```

```python
df.head()   
```

Removing the unnecssary columns


```python
columns_to_drop = [
    'magType',
    'nst',
    'gap',
    'dmin',
    'rms',
    'net',
    'id',
    'updated',
    'horizontalError',
    'depthError',
    'magError',
    'magNst',
    'locationSource',
    'magSource'
]

df = df.drop(columns=columns_to_drop)

print(df.head())
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
df['time'] = pd.to_datetime(df['time'])

df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month
```

```python
df
```

```python
avg_mag = np.mean(df['mag'])
max_mag = np.max(df['mag'])
min_mag = np.min(df['mag'])

print("Average Magnitude:", avg_mag)
print("Minimum Magnitude",min_mag)
print("Maximum Magnitude:", max_mag)
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['mag'],
    bins=20,
    color='orange',
    edgecolor='black'
)

plt.title('Magnitude Distribution')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')

plt.show()
```

year wise Eartquake counts 

```python
yearly_counts = df.groupby('year').size()

print(yearly_counts)
```

```python

yearly = df.groupby('year').size()

yearly.plot(
    kind='line',
    figsize=(10,5),
    title='Earthquakes Per Year'
)
plt.ylabel('Count')
plt.show()
```

Country/Region Analysis

```python
top_regions = df['place'].value_counts().head(10)

print(top_regions)
```

```python
top_places = df['place'].value_counts().head(10)

top_places.plot(
    kind='barh',
    figsize=(10,6),
    color='red'
)

plt.title('Top 10 Earthquake Locations')

plt.show()
```

Depth Analysis

```python
depth_stats = df['depth'].describe()

print(depth_stats)
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['depth'],
    bins=20,
    color='skyblue'
)

plt.title('Depth Distribution')
plt.xlabel('Depth')
plt.ylabel('Count')

plt.show()
```

Correlation Analysis

```python

plt.figure(figsize=(8,6))

sns.heatmap(
    df[['latitude', 'longitude', 'depth', 'mag']].corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Earthquake Feature Correlation')
plt.show()
```

```python
correlation = df[['mag', 'depth']].corr()

print(correlation)
```

Risk Classification

```python
def risk_level(mag):
    if mag >= 7:
        return "High"
    elif mag >= 5:
        return "Medium"
    else:
        return "Low"

df.loc[:, 'risk'] = df['mag'].apply(risk_level)
print(df[['mag', 'risk']].head())
```

```python
risk_counts = df['risk'].value_counts()

plt.figure(figsize=(5,5))

plt.pie(
    risk_counts,
    labels=risk_counts.index,
    autopct='%1.1f%%'
)

plt.title('Earthquake Risk Categories')

plt.show()
```

Hotspot Detection

```python
hotspots = df.groupby(['latitude', 'longitude']).size()

print(hotspots.sort_values(ascending=False).head())
```

Monthly Earthquake Trend

```python
monthly = df.groupby('month').size()

monthly.plot(
    kind='bar',
    color='green',
    figsize=(10,5)
)

plt.title('Monthly Earthquake Frequency')
plt.xlabel('Month')
plt.ylabel('Count')

plt.show()
print(monthly)
```

```python
df['country'] = df['place'].str.split(',').str[-1].str.strip()
df['country'].value_counts().head(10)
```

```python

top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_countries.values,
    y=top_countries.index,
    palette='viridis'
)

plt.title('Top 10 Countries by Earthquake Count', fontsize=14)
plt.xlabel('Number of Earthquakes')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```
